```python
# 5. Marketing team wants to know the difference in spending habits
df.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# 6. Create a spending flagfor each user
```

```python
df['spending'] = df.groupby(['user_id'])['prices'].transform(np.mean)
df.loc[df['spending'] < 10, 'spending_flag'] = 'Low Spender'
df.loc[df['spending'] >= 10, 'spending_flag'] = 'High Spender'
df['spending_flag'].value_counts(dropna=False)
```

```python
# 7. Create an Order frequency flag that marks the regularity of a users ordering behaviour
df['frequency'] = df.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df.loc[df['frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
df.loc[df['frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
df.loc[(df['frequency'] > 10) & (df['frequency'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```
